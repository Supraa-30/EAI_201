```python
import pandas as pd
import time

def standings_scraper():

    years = ['2022','2018', '2014', '2010','2006']
    all_data = {}

    for year in years:
        url = f"https://fbref.com/en/comps/1/{year}/{year}-World-Cup-Stats"
        print(f"\n=== Scraping {year} World Cup ===")

        try:
            # Read all tables from the page
            tables = pd.read_html(url)
            print(f"Found {len(tables)} tables")

            found_table = False
            for i, table in enumerate(tables):
                print(f"Table {i}: Shape {table.shape}")
                print(f"Columns: {list(table.columns)[:5]}...")  # Show first 5 columns

                # Look for tables with 25-40 rows and 10-25 columns as it suggests standings
                if (25 <= len(table) <= 40) and (10 <= len(table.columns) <= 25):
                    print(f"Table {i} looks like standings! Saving...")
                    all_data[year] = table
                    found_table = True
                    break

            if not found_table:
                print(f"No suitable table found for {year}")
                # Save the first table anyway for inspection
                if tables:
                    all_data[year] = tables[0]
                    print(f"Saved first table for inspection")

            time.sleep(3)

        except Exception as e:
            print(f"Error with {year}: {e}")

    # Save results
    if all_data:
        filename = 'world_cup_standings.xlsx'
        with pd.ExcelWriter(filename) as writer:
            for year, df in all_data.items():
                # Clean sheet name
                sheet_name = f"WC_{year}"[:31]
                df.to_excel(writer, sheet_name=sheet_name, index=False)
                print(f"Saved {year} data ({df.shape}) to sheet: {sheet_name}")

        print(f"\n File saved as: {filename}")

    return all_data


standings_scraper()
```
